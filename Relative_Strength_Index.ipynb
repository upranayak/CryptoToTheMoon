```python
import datetime as dt
import talib as ta
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
df = pd.read_csv('BTC_ETH_LTC_Jun2522_Jul0222_1h.csv')
```

```python
print(df.head(5))
```

```python
df['RSI_BTC']=ta.RSI(df['BTC-USD_Close'])
```

```python
fig, axs = plt.subplots(2, 1, gridspec_kw={"height_ratios": [3, 1]},figsize=(10,6))
axs[0].plot(df['BTC-USD_Close'])
axs[1].axhline(y=70, color="r", linestyle="--")
axs[1].axhline(y=30, color="g", linestyle="--")
axs[1].plot(df['RSI_BTC'], color="orange")
```
